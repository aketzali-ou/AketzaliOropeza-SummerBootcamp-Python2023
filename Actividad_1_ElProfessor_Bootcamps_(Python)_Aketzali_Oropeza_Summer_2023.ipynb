```python
#Ejercicio 26 -  Matemáticas Discretas Pt.4

def Factorial (n):
  Resultado=1 #El multiplicación, el 1 no afecta
  for Contador in range(n): #El range PUEDE TENER HASTA 3 ELEMENTOS: límite superior, límite inferior, paso
    Resultado=Resultado+(Resultado*Contador)
    Contador=Contador+1
  return Resultado

n=int(input("Introduce un número para obtener su factorial: "))
print(f"El factorial de {n} es {Factorial(n)}")
```

```python
#Ejercicio 26 -  Matemáticas Discretas Pt.5 - Combinaciones y permutaciones

def Factorial (n):
  Resultado=1
  for Contador in range(n):
    Resultado=Resultado+(Resultado*Contador)
    Contador=Contador+1
  return Resultado

def Permutar (m,n):
  return int(Factorial(m)/Factorial(m-n))

def Combinar (m,n):
  return int(Factorial(m)/((Factorial(m-n))*Factorial(n)))

m=int(input("Población (m): "))
n=int(input("Número de elementos a elegir de la población (n): "))

print ("")
print (f"El número de PERMUTACIONES es {Permutar(m,n)}")
print (f"El número de COMBINACIONES es {Combinar(m,n)}")
```
